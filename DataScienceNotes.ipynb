# Data Science: Visualização de Dados com Python

Aprenda a construir visualizações de dados usando a linguagem de programação Python e a biblioteca Pyplot.

```python
"Hello World MatPlotLib!"
```

```python
import matplotlib.pyplot as plt
```

### Gráficos de linhas

```python
x = [1,2]
y = [2,3]
```

```python
plt.plot(x,y)
plt.show()
```

### Legendas para gráficos

```python
x = [1,2,5]
y = [2,3,7]
```

```python
plt.title("Grafico de linhas PyPlot")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.plot(x, y)
plt.show()
```

### Gráficos de Barras

```python
x = [1,2,3,4,5,6]
y = [2,3,7,6,2,10]
```

```python
plt.title("Grafico de Barras PyPlot")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.bar(x, y)
plt.show()
```

### Comparação em gráfico de barras


```python
x1 = [1,3,5,7,9,11]
y1 = [2,3,7,6,2,10]

x2 = [2,4,6,8,10,12]
y2 = [4,4,8,6,1,12]
```

```python
plt.title("Grafico de Barras PyPlot")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")

plt.legend()
plt.show()
```
