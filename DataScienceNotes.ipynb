### Estudo de caso

#### Crescimento da população brasileira 1980-2016

Neste estudo de caso, criaremos um gráfico do crescimento da população brasileira.

**Fonte:** [DataSUS](https://datasus.saude.gov.br/)

**Atenção:** os dados brutos (populacao_brasileira.csv) foram submetidos como material suplementar desta aula.

````python
import pandas as pd
````

````python
dados = pd.read_csv('data/input/populacao.csv', delimiter = ';')
dados.head()
````

````python
x = dados['ano']
y = dados['population']
````

````python
plt.title("Crescimento da população brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x100.000.000")

plt.plot(
    x, y,
    label = "Crescimento da população",
    color = "k",
    linestyle = ":"
)

plt.legend()

plt.savefig("data/output/crescimento_populacao_brasileira_line.png", dpi = 72)
````

````python
plt.title("Crescimento da população brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x100.000.000")

plt.bar(
    x, y,
    label = "Crescimento da população",
    color = "#e4e4e4",
    linestyle = ":"
)

plt.legend()

plt.savefig("data/output/crescimento_populacao_brasileira_bar.png", dpi = 72)
````

````python
plt.title("Crescimento da população brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x100.000.000")

plt.scatter(
    x, y,
    label = "Crescimento da população",
    color = "k",
    marker = "."
)

plt.legend()

plt.savefig("data/output/crescimento_populacao_brasileira_scatter.png", dpi = 72)
````

````python
plt.title("Crescimento da população brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x100.000.000")

plt.bar(
    x, y,
    label = "Crescimento da população",
    color = "#e4e4e4",
    linestyle = ":"
)

plt.plot(
    x, y,
    label = "Crescimento da população",
    color = "k",
    linestyle = "-"
)

plt.legend()

plt.savefig("data/output/crescimento_populacao_brasileira.png", dpi = 72)
````

##### Com base na análise do gráfico, em que ano a população brasileira superou 150 milhões de pessoas?

````python
alvo = 15 * (10**7) # 150.000.000
ano = None
valor = None

for i in range(len(dados["population"])):
    if int((dados["population"][i])) > alvo:
        ano = int((dados["ano"][i]))
        valor = int((dados["population"][i]))
        break

print(
    "O ano que a população brasileira superou " +
    "150 milhões de pessoas foi em " + str(ano) + " " +
    "com o valor de " + str(valor) + " pessoas."
)
````

### O que é um boxplot?

```
Boxplot (diagrama de caixa) é uma
técnica de visualização de dados
em que representa a variação de dados
por meio de quartis.
```

O retângulo central concentra 50% dos dados plotados. A linha ao centro indica a mediana. Os círculos representam os outlines (valores que destoam muito dos outros valores apresentados).


````python
import random
````

````python
x = []
for i in range(100):
    x.append(random.randint(1,100))
````

````python
plt.boxplot(x)
plt.title("Gráfico Boxplot")
plt.show()
````

#### Bioinformática - comparando genomas

```
Neste estudo de caso, faremos a
comparação entre duas sequências de
DNA: (1) ser humano; vs. (2) bactéria.
```

Estrutura do DNA. Fonte: https://se.wikipedia.org/wiki/Fiila:Dna-base-flipping.svg

```
DNA é uma molécula presente em todos os seres vivos, que é responsável por armazenar as características hereditárias. Ela é composta por sequências de nucleotídeos, que podem de quatro tipos: adenina, timina, citosina ou guanina.
```

"Computacionalmente" falando podemos representá-los através de 4 letras: A, T, C ou G.

Neste estudo de caso, queremos avaliar se estruturas com funções parecidas (estamos usando sequências de RNA ribossomal) de organismos diferentes têm diferenças. Para isso vamos avaliar a quantidade de pares de nucleotídeos.

```
Dados utilizados:
* anexados como recursos desta aula (baixe o arquivo .zip)
```

```
Human 18S rRNA gene, complete
```

```
Escherichia coli strain U 5/41 16S ribosomal RNA gene, partial sequence
```
Fonte: https://www.ncbi.nlm.nih.gov/nuccore/NR_024570.1?report=fasta

#### Resultado Final

Como resultado final, geraremos uma figura que compara os DNAs visualmente.


````python
dna_bacteria_entrada = open(
    "data/input/bacteria.fasta"
).read().replace("\n","")

dna_bacteria_saida = open("data/output/bacteria.html","w")
````

````python
cont = {}
nt = ["A", "T", "C", "G"]
````

````python
for i in nt:
    for j in nt:
        cont[i + j] = 0
````

````python
for i in range(len(dna_bacteria_entrada) - 1):
    cont[dna_bacteria_entrada[i] + dna_bacteria_entrada[i + 1]] += 1
````

````python
i = 1
combinacoes = list(cont.keys())
combinacoes.sort()
for combinacao in combinacoes:
    transparencia = cont[combinacao] / max(cont.values())
    dna_bacteria_saida.write(
        '<div style = "width:100px; border: 1px solid #111; height: 100px; float: left; ' +
        'background-color: rgba(0, 0, 255,' +
        str(transparencia) +
        '); color: #ffffff">' +
        str(combinacao) +
        '</div>\n'
    )
    if i % 4 == 0:
        dna_bacteria_saida.write(
            '<div style = "clear: both;"></div>\n'
        )
    i = i + 1
dna_bacteria_saida.close()
````
